```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['n_autopista'] = le.fit_transform(df['autopista'])
df['n_condiciones_meteorologicas'] = le.fit_transform(df['condiciones_meteorologicas'])
df['n_superficie_de_la_via'] = le.fit_transform(df['superficie_de_la_via'])
df['n_tipo_de_siniestro'] = le.fit_transform(df['tipo_de_siniestro'])
```

```python
data = df[['hora', 'n_autopista', 'n_condiciones_meteorologicas', 'n_superficie_de_la_via', 'n_tipo_de_siniestro', 'moto_o_liviano_involucrado', 'camion_o_bus_involucrado']]
target_lesionados = df['lesionados'] = pd.to_numeric(df['lesionados']).apply(lambda x: 1 if x > 0 else 0)
```

```python
data.shape
```

```python
target_lesionados.shape
```

```python
trainset_data = numpy.array(data[:5000])
trainset_target_lesionados = target_lesionados[:5000]

testset_data = numpy.array(data[5000:])
testset_target_lesionados = target_lesionados[5000:]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
```

```python
tree_clf.fit(trainset_data, trainset_target_lesionados)
```

```python
import graphviz
from sklearn.tree import export_graphviz

feature_names = ['hora', 'autopista', 'condiciones_meteorologicas', 'superficie_de_la_via', 'tipo_de_siniestro', 'moto_o_liviano_involucrado', 'camion_o_bus_involucrado']

export_graphviz(
        tree_clf,
        out_file="lesionados.dot",
        feature_names=feature_names,
        class_names='lesionado',
        filled=True
    )

with open("lesionados.dot") as f:
    dot_graph = f.read()

# remove the display(...)

display(graphviz.Source(dot_graph))
```

```python
pred_target_lesionados = tree_clf.predict(testset_data)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

species = np.array(testset_target_lesionados)
predictions = np.array(pred_target_lesionados)
confusion_matrix(species, predictions)
```
