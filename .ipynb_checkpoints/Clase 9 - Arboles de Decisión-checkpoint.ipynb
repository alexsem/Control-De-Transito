# Arboles de decisi√≥n #

````python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
````

````python
iris = load_iris()
iris.feature_names[2:]
````

````python
iris_X = iris.data[:, 2:]  # petal length and width
iris_y = iris.target

iris_X[:6], iris_y[:6]
````

````python
plt.plot(iris_X[iris_y==0,0],iris_X[iris_y==0,1], "b.")
plt.plot(iris_X[iris_y==1,0],iris_X[iris_y==1,1], "g.")
plt.plot(iris_X[iris_y==2,0],iris_X[iris_y==2,1], "r.")
````

Entrenar un arbol de decisi√≥n:

````python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(iris_X, iris_y)
````

````python
from sklearn.tree import export_graphviz
````

````python
export_graphviz(
        tree_clf,
        out_file="iris_tree.dot",
        feature_names=iris.feature_names[2:],
        class_names=iris.target_names,
        filled=True
    )
````

# Arboles de regresi√≥n

````python
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
X[:6]
````

````python
y = 4 * (X - 0.5) ** 2
y[:6]
````

````python
plt.plot(X, y, "b.")
````

````python
y = y + np.random.randn(m, 1) / 10

plt.plot(X, y, "b.")
````

````python
from sklearn.tree import DecisionTreeRegressor
````

````python
tree_reg = DecisionTreeRegressor(max_depth=2, random_state=0)
tree_reg.fit(X, y)
````

````python
xmin=0
xmax=1
x1 = np.linspace(xmin, xmax, 500).reshape(-1, 1)
(x1[:6],'...', x1[-6:])
````

````python
y_pred = tree_reg.predict(x1)
````

````python
plt.plot(X, y, "b.")
plt.plot(x1, y_pred, "r-")
plt.show()
````

````python
tree_reg2 = DecisionTreeRegressor(random_state=0, max_depth=3)
tree_reg2.fit(X, y)
````

````python
y_pred2 = tree_reg2.predict(x1)

plt.plot(X, y, "b.")
plt.plot(x1, y_pred, "r-")
plt.show()

plt.plot(X, y, "b.")
plt.plot(x1, y_pred2, "r-")
plt.show()
````

````python
import graphviz

def show_tree(tree_clf, feature_names):
    export_graphviz(
            tree_clf,
            out_file="tree.dot",
            feature_names=feature_names,
            filled=True
        )

    with open("tree.dot") as f:
        dot_graph = f.read()

    # remove the display(...)

    return graphviz.Source(dot_graph)
````

````python
show_tree(tree_reg, ["x1"])
````

````python
tree_reg.predict(0.4)
````

````python
plt.plot(X, y, "b.")
plt.plot(x1, y_pred, "r-")
plt.axvline(x=0.4, linestyle=":")
plt.show()
````

````python
show_tree(tree_reg2, ["x1"])
````

````python
tree_reg_free = DecisionTreeRegressor(random_state=0)
tree_reg_free.fit(X, y)
````

````python
y_pred_free = tree_reg_free.predict(x1)

plt.plot(X, y, "b.")
plt.plot(x1, y_pred_free, "r-")
plt.show()
````

````python
show_tree(tree_reg_free, ["x1"])
````

# Limites de decision

Ignorar esta funcion, sirve para graficar los limites de decisi√≥n.

````python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[0, 7.5, 0, 3], iris=True, legend=False, plot_training=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap, linewidth=10)
    if not iris:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris-Setosa")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris-Versicolor")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Iris-Virginica")
        plt.axis(axes)
    if iris:
        plt.xlabel("Petal length", fontsize=14)
        plt.ylabel("Petal width", fontsize=14)
    else:
        plt.xlabel(r"$x_1$", fontsize=18)
        plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
    if legend:
        plt.legend(loc="lower right", fontsize=14)
````

````python
plot_decision_boundary(tree_clf, iris_X, iris_y)
````
